workflow:
* create a new folder and open it in VScode
* create a test .ipynb and save it
* add a test command in the notebook
* open palette in vs code <with : Ctrl+Maj+P>
* create a new .venv environnement and select the python interpreter for the base (here it is 3.12.10). VSCode creates and activates the venv.
* start evaluating the test command: vscode aske to install ipykernel in a popup window. Just do it. It is necessary for runnign python inside a notebook
* select the new .venv environnement and evaluate the test cell.
* It might by necessary to restart the kernel or quit vscode
* when reopening the folder in vscode, the .venv is selected by default


%pip will install the package in the virtual environment where the current notebook kernel is running. While !pip will install the package in the base environment. If you are using Python virtual environment (as you should!), you should use %pip.
